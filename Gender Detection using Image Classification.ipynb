```python
#Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Prepare / Dataset
```

```python
import os
import cv2 

path = os.listdir('C:/Users/Anuj Singh Tomar/OneDrive/Documents/Data Science/Projects/Gender Detection/Gender_Image_Data/Training/')
classes = {'Female':0, 'Male':1}
```

```python
import cv2
X = []
Y = []
for cls in classes:
    pth ='C:/Users/Anuj Singh Tomar/OneDrive/Documents/Data Science/Projects/Gender Detection/Gender_Image_Data/Training/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j , 0)
        img = cv2.resize(img, (200,200))
        X.append(img)
        Y.append(classes[cls])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_updated, Y, random_state=10,
                                               test_size=.20)
```

# Feature Scaling

```python
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
xtrain = xtrain/255
xtest = xtest/255
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
```

# Train the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
import warnings
warnings.filterwarnings('ignore')

lg = LogisticRegression(C=0.1)
lg.fit(xtrain, ytrain)
```

```python
sv = SVC()
sv.fit(xtrain, ytrain)
```

## Evaluation

```python
print("Training Score:", lg.score(xtrain, ytrain))
print("Testing Score:", lg.score(xtest, ytest))
```

```python
print("Training Score:", sv.score(xtrain, ytrain))
print("Testing Score:", sv.score(xtest, ytest))
```
